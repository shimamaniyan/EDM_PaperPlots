```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
```

```python
fig, ax = plt.subplots(figsize=(8,3.5))
x = ['QUEST', 'Naïve Bayes', 'CHAID', 'C&RT', 'C5.0', 'SVM']
y1 = [63, 90, 81, 79, 93, 99]
y2 = [78, 83, 84, 85, 89, 97]
y3 = [86, 88, 90, 79, 94, 98]
ax.grid(color='lightgray', linestyle='--', linewidth=0.5)
ax.plot(x, y1, label = 'M.Sc. GPA', marker= 'D', markersize= 5, linewidth=2, linestyle='-', c= 'brown')
ax.plot(x, y2, label = 'M.Sc. duration', marker= 'D', markersize= 5, linewidth=2, linestyle='-', c= 'gray')
ax.plot(x, y3, label = 'Count of fail marks', marker= 'D', markersize= 5, linewidth=2, linestyle='-', c= 'teal')
ax.legend(loc='lower center',bbox_to_anchor=(0.5,0.01), ncol= 3)
fmt = '%.0f%%' # Define the format
yticks = mtick.FormatStrFormatter(fmt) # Create the formatter
ax.yaxis.set_major_formatter(yticks)
for spine in ax.spines.values():
    spine.set_edgecolor('gray')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6,3.5))
labels = ['Innovative Techniques','Predictive Modeling', 'Early Forecasting',
         'Analyzing Educational Institutions', 'Analyzing Student Behaviour',
         'Review Papers']
values = [0.23, 0.32, 0.04, 0.10, 0.26, 0.05]
colors = ['teal', 'teal', 'brown', 'teal', 'teal', 'teal']
rects = ax.barh(labels, values, color= colors)
ax.bar_label(rects, labels=['{:.0%}'.format(val) for val in values])
ax.invert_yaxis()
plt.show()
```
